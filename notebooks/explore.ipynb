```python
import sys
sys.path.append('..')
from src.data.dataset import UrbanIssuesDataset
from collections import Counter
import pandas as pd

# No need for transforms if we are just inspecting labels
train_dataset = UrbanIssuesDataset(root_dir='../data', split='train', transform=None)
val_dataset = UrbanIssuesDataset(root_dir='../data', split='valid', transform=None)
test_dataset = UrbanIssuesDataset(root_dir='../data', split='test', transform=None)

# Access samples directly to avoid loading images
train_labels = [label for _, label in train_dataset.samples]
val_labels = [label for _, label in val_dataset.samples]
test_labels = [label for _, label in test_dataset.samples]

train_counts = Counter(train_labels)
val_counts = Counter(val_labels)
test_counts = Counter(test_labels)

# Get class names from the dataset
classes = train_dataset.classes
all_indices = sorted(list(set(train_counts.keys()) | set(val_counts.keys()) | set(test_counts.keys())))

class_counts_df = pd.DataFrame({
    'Class ID': all_indices,
    'Class Name': [classes[i] for i in all_indices],
    'Train': [train_counts.get(i, 0) for i in all_indices],
    'Validation': [val_counts.get(i, 0) for i in all_indices],
    'Test': [test_counts.get(i, 0) for i in all_indices],
})
print(class_counts_df)
```
